```python
import yfinance as yf

def get_historical_stock_price(ticker: str, interval: str,  from_date: str, to_date: str):
    stock = yf.Ticker(ticker)
    data = stock.history(interval=interval, start=from_date,  end=to_date)
    if data.empty:
        raise Exception(f"No data found for ticker {ticker} in the specified date range.")
    #return data[['Open', 'High', 'Low', 'Close', 'Volume']]
    return data
```

```python
msft_hist_data = get_historical_stock_price(ticker="MSFT", interval="1mo", from_date="2000-01-01", to_date="2024-01-10")
msft_hist_data
```
