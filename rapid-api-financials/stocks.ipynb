```python
import http.client

conn = http.client.HTTPSConnection("macrotrends-finance1.p.rapidapi.com")

headers = {
    'x-rapidapi-key': "287c5058a7mshea84886f4cf6c10p1010f4jsna4227c798ad5",
    'x-rapidapi-host': "macrotrends-finance1.p.rapidapi.com"
}

conn.request("GET", "/financial-statements/msft", headers=headers)

res = conn.getresponse()
data_msft = res.read()

print(data_msft.decode("utf-8"))
```

```python
data_msft
```

```python
import json
import pandas as pd

data_json = json.loads(data_msft)
```

```python
# Extract the balance sheet data
balance_sheet = data_json['balance-sheet']
#balance_sheet

# Create a DataFrame from the balance sheet
df = pd.DataFrame.from_dict(balance_sheet, orient='index')
df.head()
```

```python
# Rename the columns using the "Date" key for better readability
df.columns = df.loc['Date']
df = df.drop('Date')
```

```python
df.transpose()
```

```python
# Extract the balance sheet data
msft_income_statement = data_json['income-statement']
#msft_income_statement

# Create a DataFrame from the balance sheet
df_ic_msft = pd.DataFrame.from_dict(msft_income_statement, orient='index')

# Rename the columns using the "Date" key for better readability
df_ic_msft.columns = df_ic_msft.loc['Date']
df_ic_msft = df_ic_msft.drop('Date')
df_ic_msft = df_ic_msft.transpose()
df_ic_msft.head()
```
