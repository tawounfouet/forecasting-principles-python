```python
import requests

def get_country_indicator_data(country_code, indicator_code, start_year, end_year):
    """
    Récupère les données d'indicateur d'un pays donné sur une plage de dates.
    
    :param country_code: Code ISO du pays (ex: 'FR' pour la France)
    :param indicator_code: Code de l'indicateur (ex: 'SP.POP.TOTL' pour la population totale)
    :param start_year: Année de début (ex: 2000)
    :param end_year: Année de fin (ex: 2023)
    :return: Données des indicateurs sous forme JSON
    """
    url = f"http://api.worldbank.org/v2/country/{country_code}/indicator/{indicator_code}"
    params = {
        'date': f'{start_year}:{end_year}',
        'format': 'json'
    }
    
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        data = response.json()
        # Les résultats sont dans la deuxième partie de la réponse JSON
        if len(data) > 1:
            return data[1]
        else:
            return None
    else:
        print(f"Erreur lors de la requête: {response.status_code}")
        return None

# Exemple d'utilisation
country_code = 'FR'  # Code ISO pour la France
indicator_code = 'SP.POP.TOTL'  # Population totale
start_year = 2000
end_year = 2023

data = get_country_indicator_data(country_code, indicator_code, start_year, end_year)


```

```python
import requests
import pandas as pd

def get_country_indicator_data(country_code, indicator_code, start_year, end_year):
    url = f"http://api.worldbank.org/v2/country/{country_code}/indicator/{indicator_code}"
    params = {
        "date": f"{start_year}:{end_year}",
        "format": "json",
        "per_page": 1000
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        if data and len(data) > 1:
            return data[1]  # les données sont dans le deuxième élément de la réponse
    return None

# Liste des pays et leurs codes (exemple avec Etats-Unis, Chine, France, Angleterre, Allemagne, et Cameroun)
countries = {
    "Etats-Unis": "US",
    "Chine": "CN",
    "France": 'FR',
    "Royaume-Uni": "GB",
    "Allemagne": 'DE',
    "Cameroun": 'CM'
}

# Liste pour stocker les résultats
all_data = []

# Indicateurs à récupérer
indicators = {
    "PIB par habitant": 'NY.GDP.PCAP.CD',
    "PNB": 'NY.GNP.MKTP.CD',
    "PNB par habitant": 'NY.GNP.PCAP.CD',
    "Exportations": 'NE.EXP.GNFS.CD',
    "Importations": 'NE.IMP.GNFS.CD',
    "Dette externe": 'DT.DOD.DECT.CD',
    "Taux de croissance du PIB": 'NY.GDP.MKTP.KD.ZG',
    "Population totale": 'SP.POP.TOTL',
    "Taux de croissance de la population": 'SP.POP.GROW',
    "Indice de fécondité": 'SP.DYN.TFRT.IN',
    "Gini": 'SI.POV.GINI',
    "Inflation (Indice des prix à la consommation)": 'FP.CPI.TOTL.ZG',
    "Taux d'intérêt réel": 'FR.INR.RINR',
    "Taux de chômage": 'SL.UEM.TOTL.ZS'
}

start_year = 2000
end_year = 2023

# Parcourir les pays et les indicateurs pour récupérer les données
for country_name, country_code in countries.items():
    for name, indicator_code in indicators.items():
        data = get_country_indicator_data(country_code, indicator_code, start_year, end_year)
        if data:
            for entry in data:
                all_data.append({
                    "Pays": country_name,
                    "Code pays": country_code,
                    "Indicateur": name,
                    "Code indicateur": indicator_code,
                    "Année": entry['date'],
                    "Valeur": entry['value']
                })

# Convertir la liste en DataFrame
df_countries = pd.DataFrame(all_data)


# Affichage du DataFrame
print(df_countries.shape)
```

```python
import requests
import pandas as pd

# Fonction pour récupérer les données d'un indicateur pour un pays spécifique sur une année donnée
def get_indicator_for_country_year(country_code, indicator_code, year):
    url = f"http://api.worldbank.org/v2/country/{country_code}/indicator/{indicator_code}"
    params = {
        "date": year,
        "format": "json",
        "per_page": 1000
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        if data and len(data) > 1 and data[1]:
            return data[1][0]['value']  # Récupérer la valeur pour cette année
    return None  # Si les données ne sont pas disponibles

# Chargement du DataFrame de pays avec les codes ISO
# Exemple : df_countries = pd.read_csv('countries.csv')
# Le DataFrame doit avoir une colonne 'Code ISO' ou similaire


```

```python
# Exemple d'un DataFrame simple avec les codes ISO de certains pays
data = {'Pays': ['Etats-Unis', 'Chine', 'France', 'Allemagne', 'Cameroun'],
        'Code ISO': ['US', 'CN', 'FR', 'DE', 'CM']}
data_countries = pd.DataFrame(data)

data_countries
```

```python
# Indicateur à ajouter (exemple: PIB par habitant)
indicator_code = 'NY.GDP.PCAP.CD'
year = '2022'

# Ajouter une nouvelle colonne avec les valeurs de l'indicateur
data_countries[f'PIB par habitant {year}'] = data_countries['Code ISO'].apply(
    lambda code: get_indicator_for_country_year(code, indicator_code, year)
)

# Afficher le DataFrame avec la nouvelle colonne
data_countries
```
