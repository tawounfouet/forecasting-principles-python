```python
import yfinance as yf

def get_historical_stock_price(ticker: str, start_date: str, end_date: str):
    stock = yf.Ticker(ticker)
    data = stock.history(start=start_date, end=end_date)
    if data.empty:
        raise Exception(f"No data found for ticker {ticker} in the specified date range.")
    return data[['Open', 'High', 'Low', 'Close', 'Volume']]
```

```python
# Example usage
try:
    data = get_historical_stock_price("AAPL", "2024-01-01", "2024-01-10")
    display(data)
except Exception as e:
    print(e)
```

```python
import yfinance as yf
from dataclasses import dataclass
from datetime import datetime
import pytz

@dataclass
class YahooFinanceClient:
    ticker: str = "AAPL"
    interval: str = "1d"  # Options: '1m', '5m', '15m', '1d', '1wk', '1mo'
    from_date: str = "2024-01-09"
    to_date: str = "2024-01-09"
    
    def fetch_data(self):
        # Download historical data using yfinance
        stock = yf.Ticker(self.ticker)
        data = stock.history(interval=self.interval, start=self.from_date, end=self.to_date)
        if data.empty:
            raise Exception(f"Ticker {self.ticker} has no results for the specified date range.")
        return data

    def transform_result(self, data):
        # Transform data to match desired output format
        dataset = []
        for index, row in data.iterrows():
            utc_timestamp = index.tz_localize(pytz.utc) if not index.tzinfo else index
            dataset.append({
                'open_price': row['Open'],
                'close_price': row['Close'],
                'high_price': row['High'],
                'low_price': row['Low'],
                'volume': row['Volume'],
                'time': utc_timestamp,
            })
        return dataset

    def get_stock_data(self):
        data = self.fetch_data()
        return self.transform_result(data)
```

```python
client = YahooFinanceClient(ticker="LVMH", interval="1d", from_date="2024-01-01", to_date="2024-01-10")
stock_data = client.get_stock_data()
```
